```python
# Install library PuLP (jika belum terinstall)
!pip install pulp

# Import library
from pulp import LpProblem, LpVariable, LpMaximize, value

# ---------------------------------------------------------
# 1. MEMBUAT MODEL OPTIMASI
# ---------------------------------------------------------
model = LpProblem("Optimasi_Produksi_Jus", LpMaximize)

# ---------------------------------------------------------
# 2. VARIABEL KEPUTUSAN
# x = jumlah jus apel
# y = jumlah jus mangga
# ---------------------------------------------------------
x = LpVariable("Jus_Apel", lowBound=0, cat="Integer")
y = LpVariable("Jus_Mangga", lowBound=0, cat="Integer")

# ---------------------------------------------------------
# 3. FUNGSI OBJEKTIF
# Profit = 6000x + 5000y
# ---------------------------------------------------------
model += 6000*x + 5000*y, "Total_Keuntungan"

# ---------------------------------------------------------
# 4. KENDALA
# Bahan baku:
# - Jus apel butuh 2 apel → 2x ≤ 40
# - Jus mangga butuh 1 mangga → y ≤ 30
# ---------------------------------------------------------
model += 2*x <= 40, "Batas_Apel"
model += y <= 30, "Batas_Mangga"

# ---------------------------------------------------------
# Waktu produksi:
# - Apel: 3 menit per gelas
# - Mangga: 2 menit per gelas
# Total waktu 60 menit
# 3x + 2y ≤ 60
# ---------------------------------------------------------
model += 3*x + 2*y <= 60, "Batas_Waktu"

# ---------------------------------------------------------
# 5. MENYELESAIKAN MODEL
# ---------------------------------------------------------
model.solve()

# ---------------------------------------------------------
# 6. MENAMPILKAN HASIL
# ---------------------------------------------------------
print("===== HASIL OPTIMASI PRODUKSI JUS =====")
print("Status:", model.status)
print("Jumlah Jus Apel (x):", value(x))
print("Jumlah Jus Mangga (y):", value(y))
print("Total Keuntungan Maksimum: Rp", value(model.objective))
```
